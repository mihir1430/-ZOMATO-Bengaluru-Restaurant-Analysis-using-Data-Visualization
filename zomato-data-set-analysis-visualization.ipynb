```python
import kagglehub
import os

# Download dataset
path = kagglehub.dataset_download("himanshupoddar/zomato-bangalore-restaurants")
print("Path to dataset files:", path)

# List files inside the dataset folder
print(os.listdir(path))

# Load the CSV file (the dataset has "zomato.csv")
csv_path = os.path.join(path, "zomato.csv")
df = pd.read_csv(csv_path, encoding="latin-1")  # encoding needed for special characters
print(df.head())
```

# deleting irrelevent columns 

```python
df = df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis = 1) # axis =1 column wise , axis = 0, row wise
df.head()
```

```python
df.info()
```

## Dropping Duplicates

```python
df.drop_duplicates(inplace = True)
df.shape
```

## Cleaning Rate Column

```python
df['rate'].unique()
```

## Removing "NEW" ,  "-" and "/5" from Rate Column

```python
# converting string and mixed data to cleaned float 
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

## Filling Null Values in Rate Column with Mean

```python
df['rate'].fillna(df['rate'].mean(), inplace = True)
df['rate'].isnull().sum()
```

```python
df.info()
```

## Dropping Null Values

```python
df.dropna(inplace = True)
df.head()
```

# renaming the column for betteer visibility

```python
df.rename(columns = {'approx_cost(for two people)':'Cost2plates', 'listed_in(type)':'Type'}, inplace = True)
df.head()
```

# Explanation

df.rename(...)

This function is used to rename rows or column names in a DataFrame.

In your case, you are renaming two columns:

"approx_cost(for two people)" → "Cost2plates"

"listed_in(type)" → "Type"

 This is mainly done because the original column names have spaces, parentheses, and special characters which make them hard to type in code.

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

## Listed in(city) and location, both are there, lets keep only one.

```python
df = df.drop(['listed_in(city)'], axis = 1)
```

```python
df['Cost2plates'].unique()
```

## Removing , from Cost2Plates Column

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
    
df['Cost2plates'] = df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

```python
df.head()
```

## Cleaning Rest Type Column

```python
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types
```

```python
# according to the azbove code , we found that quick bites is the famous resturent in the dataset 
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

## Making Rest Types less than 1000 in frequency as others

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
        
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
        
```

## Cleaning Location Column

```python
location = df['location'].value_counts(ascending  = False)

location_lessthan300 = location[location<300]



def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
        
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

## Cleaning Cuisines Column

```python
cuisines = df['cuisines'].value_counts(ascending  = False)
cuisines_lessthan100 = cuisines[cuisines<100]

def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
        
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

```python
df.head()
```

## **Data is Clean, Lets jump to Visualization**

## Count Plot of Various Locations

```python
plt.figure(figsize=(16,10))
ax = sns.countplot(df['location'])
plt.xticks(rotation=90)

# Add percentages on top of bars
total = len(df['location'])
for p in ax.patches:
    count = p.get_height()
    percentage = f'{100 * count / total:.1f}%'   # format with 1 decimal
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='bottom', fontsize=9, rotation=90)
```

## Visualizing Types of Restaurents vs Rate 

```python
plt.figure(figsize=(14,8))
ax = sns.boxplot(x='Type', y='rate', data=df, palette='inferno')

# Annotate median values
medians = df.groupby(['Type'])['rate'].median().values
for i, median in enumerate(medians):
    ax.text(i, median + 0.05, str(round(median,2)), 
            horizontalalignment='center', color='black', weight='semibold')

plt.xticks(rotation=45)
plt.title('Ratings Distribution by Restaurant Type')
plt.show()
```

## No. of Votes, Location Wise

```python
df4 = df[['location', 'votes']] # Keeps only the location and votes columns.
df4.drop_duplicates() # Removes any duplicate rows so that each restaurant vote is counted once.
                      # (Note: drop_duplicates() without inplace=True doesn’t modify df4 unless you assign it back.)
df5 = df4.groupby(['location'])['votes'].sum() # Calculates the total votes per location by summing all votes for restaurants in that location.
df5 = df5.to_frame() #Converts the grouped series to a DataFrame for easier manipulation.
df5 = df5.sort_values('votes', ascending=False) # Sorts locations by total votes, with the most popular locations at the top.
df5.head()
```

```python
plt.figure(figsize = (15,8))
sns.barplot(df5.index , df5['votes'])
plt.xticks(rotation = 90)
```

# Top 5 Votes By Locations

```python
plt.figure(figsize=(15,8))

# Create a color list: top 5 = 'orange', others = 'blue'
colors = ['orange' if i < 5 else 'blue' for i in range(len(df5))]

sns.barplot(x=df5.index, y=df5['votes'], palette=colors)
plt.xticks(rotation=90)
plt.ylabel('Total Votes')
plt.title('Total Votes by Location (Top 5 Highlighted)')
plt.show()
```
